# Plotting in Seaborn Library

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
# remove warnings
import warnings
warnings.filterwarnings('ignore')
```

### Important Points to remember:
- Data (df)
- Plot type
- axes (x/y)
- grouping (color/hue)
- Plot size
- color palletes
- plot style
- interpretation

```python
#load dataset
df = sns.load_dataset('titanic')
df['sex'].value_counts()
```

```python
# scatter plot
# plot size
plt.figure(figsize=(5,5))
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
```

```python
# scatter plot
# defining colors
colors = {'male': 'green', 'female': 'red'}
# plot size
plt.figure(figsize=(5,5))
sns.scatterplot(x='age', y='fare', data=df, hue='sex', palette=colors)
# labels
plt.xlabel('Age')
plt.ylabel('Fare')
# title
plt.title('Age vs Fare')
# saving a plot //better to save in 300dpi as 600 dpi is heay and necessary in extreme need. save in png file
plt.savefig('./outputs/Age vs Fare.png', dpi=300)
```

```python
# setting backgroud theme
sns.set_theme(style='darkgrid', font='Times New Roman')
# plot size
plt.figure(figsize=(16,9))
# labels
plt.xlabel('Age')
plt.ylabel('Fare')
# title
plt.title('Age vs Fare')
sns.scatterplot(x='age', y='fare', data=df,size='fare', hue='class')
```

- Get key insights from this plot.

```python
# now boxplot
# setting backgroud theme
sns.set_theme(style='darkgrid', font_scale=1,font='Times New Roman')
# plot size
plt.figure(figsize=(16,9))
# labels
plt.xlabel('Class')
plt.ylabel('Fare')
# title
plt.title('Age vs Fare-BoxenPlot')
sns.boxenplot(x='class', y='fare', data=df, hue='sex')
# change legend position and name
plt.legend(loc='upper right', title='Gender')
# changing x tick labels
plt.xticks(rotation=45)
# changing x tick names
plt.xticks([0, 1, 2], ['First Class', 'Second Class', 'Third Class'])
```
